# **Analisi del dataset**

```python
# Importazione delle librerie necessarie
import pandas as pd
import numpy as np

# Caricamento dataset
file_path = '/kaggle/input/marketing-and-product-performance-dataset/marketing_and_product_performance.csv'
df = pd.read_csv(file_path)

# Visualizzazione delle prime righe del dataset
print("Prime righe del dataset:")
display(df.head())

# Informazioni sul dataset
print("\nInformazioni sul dataset:")
df_info = pd.DataFrame({
    'Colonna': df.columns,
    'Tipologia': df.dtypes,
    'Valori Non Nulli': df.notnull().sum(),
    'Valori Nulli': df.isnull().sum()
})
display(df_info)

# Statistiche descrittive per le colonne numeriche
print("\nStatistiche descrittive per colonne numeriche:")
statistiche = df.describe().T
statistiche['Valori Nulli'] = df.isnull().sum()
display(statistiche)

# Controllo dei valori unici per ogni colonna
print("\nValori unici per colonna:")
unique_values = df.nunique()
unique_table = pd.DataFrame({'Colonna': unique_values.index, 'Valori Unici': unique_values.values})
display(unique_table)
```

# **Pulizia dei dati**

```python
# Controllo dei valori mancanti
print("\nValori mancanti per colonna:")
missing_values = df.isnull().sum()
missing_table = pd.DataFrame({'Colonna': df.columns, 'Valori Mancanti': missing_values})
display(missing_table)

# Controllo e rimozione dei duplicati
duplicati = df.duplicated().sum()
print(f"\nNumero di duplicati: {duplicati}")
print(f"Righe iniziali nel dataset: {len(df)}")
df = df.drop_duplicates()
print(f"Righe dopo la rimozione dei duplicati: {len(df)}")

# Verifica e correzione dei tipi di colonna
df['Budget'] = df['Budget'].astype(float)
df['Revenue_Generated'] = df['Revenue_Generated'].astype(float)
df['Subscription_Length'] = df['Subscription_Length'].astype(int)
df['Customer_Satisfaction_Post_Refund'] = df['Customer_Satisfaction_Post_Refund'].astype(int)

# Rimozione delle colonne non necessarie
colonne_da_rimuovere = ['Flash_Sale_ID', 'Bundle_ID', 'Bundle_Price']
df = df.drop(columns=colonne_da_rimuovere)
print(f"Colonne rimosse: {colonne_da_rimuovere}")

# Creazione di nuove colonne utili
df['CPC'] = df['Budget'] / df['Clicks'].replace(0, float('nan'))  # Costo per clic
df['CPA'] = df['Budget'] / df['Conversions'].replace(0, float('nan'))  # Costo per acquisizione
df['Profit'] = df['Revenue_Generated'] - df['Budget']  # Utile netto

# Controllo dei valori mancanti dopo le modifiche
missing_values_after = df.isnull().sum().sum()
print(f"\nValori mancanti nel dataset dopo la pulizia: {missing_values_after}")

# Calcolo del punteggio di interazione: Clicks + Units_Sold
df['Interaction_Score'] = df['Clicks'] + df['Units_Sold']

# Ordinare per punteggio di interazione decrescente e seleziona i primi 300 campioni
df = df.sort_values(by='Interaction_Score', ascending=False).head(300)

# Rimozione della colonna temporanea "Interaction_Score"
df = df.drop(columns=['Interaction_Score'])

# Salva il dataset ridotto
output_path = '/kaggle/working/reduced_dataset.csv'
df.to_csv(output_path, index=False)
print("Dataset ridotto salvato con successo in:", output_path)

# Controllo delle colonne finali
print(df.columns)
display(df.head())
```

##  Analizzare come l'allocazione del budget influisce su clic, conversioni e ricavi

```python
# Raggruppamento per campagne e calcolo delle metriche
campaign_metrics = df.groupby('Campaign_ID').agg({
    'Budget': 'sum',
    'Clicks': 'sum',
    'Conversions': 'sum',
    'Revenue_Generated': 'sum'
}).reset_index()

# Aggiunta di metriche calcolate
campaign_metrics['CTR'] = (campaign_metrics['Clicks'] / campaign_metrics['Budget']) * 100  # Click-Through Rate (%)
campaign_metrics['CPC'] = campaign_metrics['Budget'] / campaign_metrics['Clicks']  # Costo per clic ($)
campaign_metrics['CPA'] = campaign_metrics['Budget'] / campaign_metrics['Conversions']  # Costo per acquisizione ($)
campaign_metrics['ROI'] = campaign_metrics['Revenue_Generated'] / campaign_metrics['Budget']  # Return on Investment

# Visualizzare le metriche principali per campagna
print("Metriche per Campagna:")
display(campaign_metrics.head())
```

## Clusterizzazione dei clienti con k-means

```python
# Clusterizzazione con il k-means
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Selezionare le colonne rilevanti per il clustering
features = df[['Subscription_Length', 'Profit']].dropna()

# Normalizzare i dati per migliorare il clustering
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Applicazione del clustering K-means
kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')
kmeans.fit(features_scaled)

df['Cluster'] = kmeans.labels_

# Centroidi
centroids = kmeans.cluster_centers_

# Visualizzazione dei cluster
plt.figure(figsize=(6, 4))
sns.scatterplot(
    x=features_scaled[:, 0],
    y=features_scaled[:, 1],
    hue=kmeans.labels_,
    palette='Set2',
    s=100
)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', color='red', s=200, label='Centroidi')
plt.title('Clusterizzazione K-means dei Clienti', fontsize=14)
plt.xlabel('Lunghezza Abbonamento', fontsize=12)
plt.ylabel('Profitto', fontsize=12)
plt.legend(title='Cluster', loc='upper right')
plt.grid(alpha=0.4, linestyle='--')
plt.tight_layout()

# Centroidi 
centroids_original = scaler.inverse_transform(centroids)
print("Centroidi (valori originali):")
print(centroids_original)

# Grafico
for cluster in df['Cluster'].unique():
    # Filter the main DataFrame based on the cluster
    cluster_data = df[df['Cluster'] == cluster]
    # Calculate customer segments for the cluster
    cluster_segments = cluster_data[['Customer_ID', 'Subscription_Length', 'Profit']].groupby('Customer_ID').mean()
    correlation_matrix = cluster_segments[['Subscription_Length', 'Profit']].corr()

    plt.figure(figsize=(6, 4))
    sns.heatmap(
        correlation_matrix,
        annot=True,
        cmap='viridis',
        fmt='.2f',
        linewidths=0.5
    )
    plt.title(f'Matrice di Correlazione per il Cluster {cluster}', fontsize=16)
    plt.show()
```

## Analisi della silhouette

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm

# Selezione delle variabili dal dataset
features = df[['Subscription_Length', 'Profit']].dropna()

# Normalizzazione dei dati
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Gamma di cluster da testare
range_n_clusters = [2, 3, 4, 5, 6]

for n_clusters in range_n_clusters:
    # Creazione del KMeans per il numero di cluster specificato
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(20, 6)

    # Configurazione dela prima subplot per il silhouette plot
    ax1.set_xlim([-0.1, 1])
    ax1.set_ylim([0, len(features_scaled) + (n_clusters + 1) * 10])
    
# Inizializzazione del modello KMeans
    clusterer = KMeans(n_clusters=n_clusters, n_init="auto", random_state=10)
    cluster_labels = clusterer.fit_predict(features_scaled)

    # Calcolo del silhouette score
    silhouette_avg = silhouette_score(features_scaled, cluster_labels)
    print(f"Per n_clusters = {n_clusters}, il silhouette_score medio : {silhouette_avg}")

    # Calcolo dei valori silhouette per ciascun punto
    sample_silhouette_values = silhouette_samples(features_scaled, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Raggruppare i valori silhouette per ciascun cluster e li ordina
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i
       
        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(
            np.arange(y_lower, y_upper),
            0,
            ith_cluster_silhouette_values,
            facecolor=color,
            edgecolor=color,
            alpha=0.7,
        )

        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Aggiornare y_lower per il cluster successivo
        y_lower = y_upper + 10

    ax1.set_title("Silhouette plot per i diversi cluster")
    ax1.set_xlabel("Valori del coefficiente di silhouette")
    ax1.set_ylabel("Etichetta del cluster")

    # Linea verticale per il silhouette score medio
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")
    
    # Configurazione della seconda subplot per i cluster
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(
        features_scaled[:, 0], features_scaled[:, 1], marker=".", s=30, lw=0, alpha=0.7, c=colors, edgecolor="k"
    )

    # Etichette per i centroidi
    centers = clusterer.cluster_centers_
    ax2.scatter(
        centers[:, 0], centers[:, 1], marker="o", c="white", alpha=1, s=200, edgecolor="k"
    )

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker=f"${i}$", alpha=1, s=50, edgecolor="k")

    ax2.set_title("Visualizzazione dei dati clusterizzati")
    ax2.set_xlabel("Prima caratteristica normalizzata")
    ax2.set_ylabel("Seconda caratteristica normalizzata")

    plt.suptitle(
        f"Analisi Silhouette per KMeans con n_clusters = {n_clusters}",
        fontsize=14
    )
    plt.show()
```

## Classificazione  

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report

# Selezione delle feature e della variabile target
features = ['Budget', 'Clicks', 'Conversions', 'Discount_Level', 'Units_Sold', 'ROI']
target = 'Profit'

# Raggruppamento dei valori di Profit in categorie
bins = [-float('inf'), 0, df[target].quantile(0.5), float('inf')]
labels = ['Loss', 'Medium', 'High']
df['Profit_Category'] = pd.cut(df[target], bins=bins, labels=labels)

# Separare le feature e il target
X = df[features]
y = df['Profit_Category']

# Codifica della variabile target
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Normalizzazione delle feature
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Suddivisione in training e test set (split casuale senza stratificazione)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.3, random_state=42)

# Addestramento del modello Random Forest
clf = RandomForestClassifier(random_state=42, n_estimators=100)
clf.fit(X_train, y_train)

# Predizione sui dati di test
y_pred = clf.predict(X_test)

# Valutazione del modello
report = classification_report(y_test, y_pred, target_names=label_encoder.classes_)
print(report)

from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

# Binarizzare le etichette per il calcolo OvR
y_test_binarized = label_binarize(y_test, classes=[0, 1, 2])
y_pred_proba = clf.predict_proba(X_test)

# Calcolo della ROC per ciascuna classe
fig, ax = plt.subplots()
for i, class_name in enumerate(label_encoder.classes_):
    fpr, tpr, _ = roc_curve(y_test_binarized[:, i], y_pred_proba[:, i])
    roc_auc = auc(fpr, tpr)
    ax.plot(fpr, tpr, label=f"ROC curve (class {class_name}) (area = {roc_auc:.2f})")

# Grafico della curva ROC
ax.plot([0, 1], [0, 1], 'k--')  
ax.set_xlim([0.0, 1.0])
ax.set_ylim([0.0, 1.05])
ax.set_title("ROC curve for multiclass (one-vs-rest)")
ax.set_xlabel("False Positive Rate")
ax.set_ylabel("True Positive Rate")
ax.legend(loc="lower right")
plt.show()
```

## Sistema di raccomandazione

```python
# Importazione librerie 
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler

# Creazione della matrice cliente-prodotto 
customer_product_matrix = df.pivot_table(index='Customer_ID', columns='Product_ID', values='Units_Sold', fill_value=0)

# Calcolo della similarit tra i prodotti
product_similarity = cosine_similarity(customer_product_matrix.T)
product_similarity_df = pd.DataFrame(
    product_similarity,
    index=customer_product_matrix.columns,
    columns=customer_product_matrix.columns
)

display("Matrice di Similarit tra Prodotti:")
display(product_similarity_df.head())  

# Normalizzare i dati per il clustering
features = df[['Subscription_Length', 'Profit', 'Units_Sold']].dropna()
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Applicazione K-Means
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)  
df['Cluster'] = kmeans.fit_predict(scaled_features)

print("Clustering completato. Distribuzione dei clienti per cluster:")
print(df['Cluster'].value_counts())

# Funzione per suggerire prodotti per un segmento specifico di clienti (cluster)
def recommend_products_for_segment(cluster_id, df, similarity_df, top_n=5):
    """
    Suggerisce prodotti per un segmento specifico di clienti (cluster).
    """
    # Filtro per i clienti appartenenti al cluster
    cluster_customers = df[df['Cluster'] == cluster_id]['Customer_ID'].unique()

    # Aggregazione dei prodotti acquistati dai clienti del cluster
    cluster_matrix = customer_product_matrix.loc[cluster_customers]
    purchased_products = cluster_matrix.sum(axis=0).sort_values(ascending=False)

    # Raccomandazione dei prodotti simili ai pi popolari
    top_products = purchased_products[purchased_products > 0].index[:top_n]
    recommendations = pd.Series(dtype=float)

    for product in top_products:
        recommendations = recommendations.add(similarity_df[product], fill_value=0)

    return recommendations.sort_values(ascending=False).head(top_n)

# Raccomandazioni per il cluster 0
recommendations_segment_0 = recommend_products_for_segment(0, df, product_similarity_df)
print("Prodotti raccomandati per il cluster 0:")
print(recommendations_segment_0)

# Raccomandazioni per il cluster 1
recommendations_segment_1 = recommend_products_for_segment(1, df, product_similarity_df)
print("Prodotti raccomandati per il cluster 1:")
print(recommendations_segment_1)

# Raccomandazioni per il cluster 2
recommendations_segment_2 = recommend_products_for_segment(2, df, product_similarity_df)
print("Prodotti raccomandati per il cluster 2:")
print(recommendations_segment_2)
```
